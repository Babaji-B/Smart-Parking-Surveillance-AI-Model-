```python
from ultralytics import YOLO
import cv2

# Load the trained model
model = YOLO('/content/best.pt')

cap = cv2.VideoCapture('/content/video.mp4')  # Path to our video file

# Getting the width and height of the frames in the video
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create VideoWriter object to save the output video
out = cv2.VideoWriter('/content/output_video.avi', cv2.VideoWriter_fourcc(*'XVID'), 30, (frame_width, frame_height))

# Setting our confidence threshold
CONFIDENCE_THRESHOLD = 0.65

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Perform prediction on the frame
    results = model(frame)

    # Access detections
    boxes = results[0].boxes

    for box in boxes:
        conf = box.conf.item()
        if conf >= CONFIDENCE_THRESHOLD:
            # Get box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            # Get label
            cls = int(box.cls[0])
            label = model.names[cls]

            # Draw the bounding box red color with thicker lines
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 3)

            cv2.putText(frame, f'{label} {conf:.2f}', (x1, y1 - 20),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)  # Font scale 1, thickness 3

    # Write the annotated frame to the output video
    out.write(frame)

# Release the video capture and writer objects
cap.release()
out.release()

print("Video processing complete! The output video is saved as 'output_video.avi'.")
```
